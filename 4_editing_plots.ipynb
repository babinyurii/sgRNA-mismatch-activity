```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
aid_df = pd.read_excel("AID.xlsx")
apobec3a_df = pd.read_excel("APOBEC3A.xlsx")
apobec3b_df = pd.read_excel("APOBEC3B.xlsx")
apobec3g_df = pd.read_excel("APOBEC3G.xlsx")
```

```python
aid_df = aid_df.sort_values(by=["mismatch_itself"])
apobec3a_df = apobec3a_df.sort_values(by=["mismatch_itself"])
apobec3b_df = apobec3b_df.sort_values(by=["mismatch_itself"])
apobec3g_df = apobec3g_df.sort_values(by=["mismatch_itself"])
```

```python
aid_df.rename(columns={"mismatch_itself" : "mismatch"}, inplace=True)
apobec3a_df.rename(columns={"mismatch_itself" : "mismatch"}, inplace=True)
apobec3b_df.rename(columns={"mismatch_itself" : "mismatch"}, inplace=True)
apobec3g_df.rename(columns={"mismatch_itself" : "mismatch"}, inplace=True)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x=aid_df["mismatch_pos"], 
                y=aid_df["predicted_activity"], 
                hue=aid_df["mismatch"],
               s=200, alpha=0.9)

plt.xticks(list(range(-20, -0, 1)))
plt.ylim(-0.07, 1.07)

plt.legend()

plt.title("AID", fontsize=18)
plt.xlabel(xlabel="mismatch position relative to PAM sequence", fontsize=15)
plt.ylabel(ylabel="predicted activity", fontsize=15)


#plt.savefig("aid_res.png", format='png', dpi=1200)

plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x=apobec3a_df["mismatch_pos"], 
                y=apobec3a_df["predicted_activity"], 
                hue=apobec3a_df["mismatch"],
               s=200, alpha=0.9)

plt.xticks(list(range(-20, -0, 1)))
plt.ylim(-0.07, 1.07)

plt.title("APOBEC3A", fontsize=18)
plt.xlabel(xlabel="mismatch position relative to PAM sequence", fontsize=15)
plt.ylabel(ylabel="predicted activity", fontsize=15)

#plt.savefig("apobec3a_res.png", format='png', dpi=1200)

plt.show()
```
