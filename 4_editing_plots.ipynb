```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pandas as pd
sns.set()
```

```python
def create_mismatch_plot(df, df_name):
    """takes df  returns plot of predicted activity
    """
    plt.figure(figsize=(12,8))
    sns.scatterplot(x=df["mismatch_pos"], 
                    y=df["predicted_activity"], 
                    hue=df["mismatch_itself"],
                   s=200, alpha=0.9)

    plt.xticks(list(range(-20, -0, 1)))
    plt.ylim(-0.07, 1.2)
    plt.legend()
    plt.title(df_name, fontsize=18)
    plt.xlabel(xlabel="mismatch position relative to PAM sequence", fontsize=15)
    plt.ylabel(ylabel="predicted activity", fontsize=15)


    plt.savefig(df_name + ".png", format='png', dpi=1200)

    #plt.show()
```

```python
for f in os.listdir("./"):
    if f.rsplit(".")[-1] == "xlsx":
        df_name = f.rsplit(".")[0]
        df = pd.read_excel(f)
        df = df.sort_values(by=["mismatch_itself"])
        create_mismatch_plot(df, df_name)
```
